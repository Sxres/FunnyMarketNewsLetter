```python
from dotenv import load_dotenv
import os 
import requests 
from supabase import create_client, Client
from datetime import datetime, timedelta, timezone
```

```python
load_dotenv()
```

```python
url: str = os.getenv("SUPABASE_URL")
key: str = os.getenv("SUPABASE_API")
news_api_key: str = os.getenv("NEWSAPI_KEY")
```

```python
tickers = ["AAPL", "NVDA", "CMG", "BBAI", "INTC", "AMD", "TSLA"]
now = datetime.now()
yesterday = now - timedelta(days=1)
```

```python
for ticker in tickers:
    url = f"https://www.alphavantage.co/query?function=NEWS_SENTIMENT&tickers={ticker}&RANGE={yesterday}&RANGE={now}&apikey={news_api_key}"
    response = requests.get(url)
    data = response.json()
```

```python
supabase: Client = create_client(url, key)
```

```python
bucket = supabase.storage.from_("StockNews")
```
