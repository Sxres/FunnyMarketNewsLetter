```python
# we just wnat 'url': and the url of each article. 'url' is a key so we can loop through and keep the url key from each article 
cleaned_ds = []
key_to_keep = 'url'
folder_path = "C:/Users/Dragos/OneDrive/Documents/FunnyMarketNewsLetter/cleandata" # change this back to linux: "/home/passdaboof/Documents/FunnyMarketNewsLetter/cleandata" or windows: C:\Users\Dragos\OneDrive\Documents\FunnyMarketNewsLetter\cleandata
```

```python
import json
for filename in os.listdir(folder_path):
    if filename.endswith(".json"):
        file_path = os.path.join(folder_path, filename)
        try:
            with open(file_path, "r", encoding="utf-8") as f:
                data = json.load(f)
            articles = data.get("feed", [])
            for article in articles:
                url = article.get(key_to_keep)
                if url:
                    cleaned_ds.append(url)
        except Exception as e:
            print(f"Error processing {filename}: {e}")

# Deduplicate while preserving order
seen = set()
unique_urls = []
for u in cleaned_ds:
    if u not in seen:
        seen.add(u)
        unique_urls.append(u)

print(f"Collected {len(cleaned_ds)} raw URLs, {len(unique_urls)} unique.")
unique_urls
```
